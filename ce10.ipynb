```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd
```

```python
penguins = pd.read_csv('penguins.csv')
```

```python
# Question 1
sns.histplot(data=penguins, x="bill_length_mm")
```

```python
# Question 2
fig = px.histogram(penguins, x="bill_length_mm")
fig.show()
```

```python
# Question 3
sns.relplot(penguins, x="bill_length_mm", y="flipper_length_mm", hue="species")
# This is a scatter plot of the penguins' bill and flipper lengths, which is then divided into species.
# Generally, Adelie penguins have shorter bill and flipper lengths compared to the two other species. Gentoo penguins have the longest flippers while Chinstrap penguins have middling flipper and bill lengths.
```
